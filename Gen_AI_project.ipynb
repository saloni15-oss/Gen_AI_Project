```python
from transformers import pipeline, set_seed
import torch
```

```python
# Set seed for reproducibility
set_seed(42)

# Load pre-trained GPT-2 model
generator = pipeline('text-generation', model='gpt2')
```

```python
prompt = "In the future, artificial intelligence will"
output = generator(prompt, max_length=60, num_return_sequences=1)

print("Generated Text:\n")
print(output[0]['generated_text'])
```

```python
prompts = [
    "Generative AI is transforming education by",
    "In a world dominated by machines, humans",
    "The secret to creativity lies in",
]

for i, prompt in enumerate(prompts):
    print(f"\nPrompt {i+1}: {prompt}")
    result = generator(prompt, max_length=50, num_return_sequences=1)
    print("Generated:", result[0]['generated_text'])
```
